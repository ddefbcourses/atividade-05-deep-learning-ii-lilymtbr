# Questão 1

Implemente uma função `load_data(seed)` que:

- carregue o dataset CIFAR-100 utilizando `tensorflow.keras.datasets.cifar100.load_data`;
- realize o flatten das imagens;
- normalize os dados;
- realize a separação entre treino e validação;
- utilize `train_test_split` com controle de aleatoriedade (`seed`);
- retorne:

```python
X_train, X_val, y_train, y_val
```

já normalizados e preparados para treinamento.

Além disso, responda:

1. Qual o formato original das imagens?
2. Quantas features cada imagem possui após o flatten?
3. Por que o flatten é necessário para uma MLP?
4. Qual a importância da normalização para o treinamento?

**Solução**:

````python
from tensorflow.keras.datasets import cifar100
from sklearn.model_selection import train_test_split

def load_data(seed=42):

    # Carregamento do dataset
    (X_train, y_train), (X_test, y_test) = cifar100.load_data()

    # Flatten das imagens
    X_train = X_train.reshape(X_train.shape[0], -1)
    X_test = X_test.reshape(X_test.shape[0], -1)

    # Normalização
    X_train = X_train.astype("float32") / 255.0
    X_test = X_test.astype("float32") / 255.0

    # Separação treino/validação
    X_train, X_val, y_train, y_val = train_test_split(
        X_train,
        y_train,
        test_size=0.2,
        random_state=seed,
        stratify=y_train
    )

    return X_train, X_val, y_train, y_val
````

1) Qual o formato original das imagens?

O dataset CIFAR-100 possui 60.000 imagens RGB no total, sendo 50.000 imagens destinadas ao conjunto de treino e 10.000 ao conjunto de teste. Cada imagem possui dimensão 32 × 32 pixels com 3 canais de cor RGB (vermelho, verde e azul). Dessa forma, o conjunto de treino possui formato (50000, 32, 32, 3).

2) Quantas features cada imagem possui após o flatten?

Após a aplicação do flatten, cada imagem passa a possuir 3072 features. Isso ocorre porque a imagem tridimensional é transformada em um vetor unidimensional através da multiplicação das dimensões: 32×32×3=3072. Assim, cada imagem passa a ser representada por um vetor contendo 3072 valores numéricos.

3) Por que o flatten é necessário para uma MLP?

O flatten é necessário porque a MLP (Multilayer Perceptron) trabalha com vetores unidimensionais como entrada. Entretanto, imagens são representadas em múltiplas dimensões, contendo altura, largura e canais de cor. Dessa forma, o flatten converte cada imagem em um vetor linear, permitindo que os dados sejam processados corretamente pelas camadas densas da rede neural.

4) Qual a importância da normalização para o treinamento?

A normalização é importante porque melhora a estabilidade e a eficiência do treinamento da rede neural. Como os valores dos pixels variam originalmente entre 0 e 255, dividir os dados por 255 faz com que os valores passem a ficar entre 0 e 1. Isso facilita o processo de otimização, melhora a convergência do gradiente e reduz instabilidades numéricas durante o treinamento, tornando o aprendizado mais rápido e consistente.

  - `activation`

A função deve retornar o modelo treinado.

Além disso, responda:

3. Por que MLPs possuem muitos parâmetros ao trabalhar com imagens?

**Solução**:

````python
from sklearn.neural_network import MLPClassifier

def train_mlp(
    X_train,
    y_train,
    activation,
    hidden_layers,
    learning_rate,
    seed
):

    model = MLPClassifier(
        hidden_layer_sizes=hidden_layers,
        activation=activation,
        learning_rate_init=learning_rate,
        random_state=seed,
        max_iter=50
    )

    model.fit(X_train, y_train.ravel())

    return model
````

1) Quantos parâmetros existem na primeira camada?

A quantidade de parâmetros da primeira camada depende do número de neurônios utilizados na camada oculta. Como cada imagem possui 3072 features após o flatten, cada neurônio da primeira camada recebe 3072 entradas. Assim, em uma arquitetura com 128 neurônios na primeira camada, a quantidade de parâmetros será: (3072×128)+128. O primeiro termo representa os pesos das conexões e o segundo representa os bias dos neurônios. Nesse caso, a camada possui 393344 parâmetros treináveis.

2) Qual a função da ativação ReLU?

A função de ativação ReLU (Rectified Linear Unit) introduz não linearidade na rede neural, permitindo que o modelo aprenda padrões mais complexos nos dados. A ReLU retorna o próprio valor da entrada quando ele é positivo e retorna zero quando o valor é negativo. Além disso, ela reduz problemas relacionados ao desaparecimento do gradiente e torna o treinamento mais rápido e eficiente.

3) Por que MLPs possuem muitos parâmetros ao trabalhar com imagens?

MLPs possuem muitos parâmetros ao trabalhar com imagens porque cada neurônio da camada oculta é conectado a todas as features da entrada. Como as imagens do CIFAR-100 possuem 3072 features após o flatten, o número de conexões cresce rapidamente, gerando centenas de milhares de pesos treináveis mesmo em arquiteturas pequenas. Isso aumenta o custo computacional, o uso de memória e a complexidade do treinamento.

# Questão 3

Implemente a função:

```python
evaluate(model, X_test, y_test)
```

Ela deve:

- realizar predições;
- calcular:
  - accuracy;
  - precision;
  - recall;
  - f1-score.

Utilize `sklearn.metrics`.

Além disso:

- interprete os resultados obtidos.

Responda:

1. O que a accuracy representa?
2. Qual a diferença entre precision e recall?
3. Em quais situações o f1-score é importante?

**Solução**:

````python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score
)

import pandas as pd

def evaluate(model, X_test, y_test):

    # Predições
    y_pred = model.predict(X_test)

    # Métricas
    results = {
        "accuracy": accuracy_score(y_test, y_pred),
        "precision": precision_score(
            y_test,
            y_pred,
            average="weighted"
        ),
        "recall": recall_score(
            y_test,
            y_pred,
            average="weighted"
        ),
        "f1_score": f1_score(
            y_test,
            y_pred,
            average="weighted"
        )
    }

    # DataFrame
    results_df = pd.DataFrame([results])

    return results_df
````

Interpretação dos resultados obtidos

Os resultados obtidos permitem avaliar o desempenho da MLP em diferentes aspectos da classificação. A accuracy indica a proporção total de previsões corretas realizadas pelo modelo. Já precision mede quantas previsões positivas estavam corretas, enquanto recall avalia quantos exemplos positivos reais foram corretamente identificados. O f1-score combina precision e recall em uma única métrica, permitindo uma avaliação mais equilibrada do desempenho do modelo.

1) O que a accuracy representa?

A accuracy representa a taxa total de acertos do modelo em relação ao número total de amostras avaliadas. Ela é calculada dividindo a quantidade de previsões corretas pelo total de exemplos analisados. Assim, uma accuracy de 0.80 indica que o modelo acertou 80% das classificações realizadas.

2) Qual a diferença entre precision e recall?

A precision mede a confiabilidade das previsões positivas feitas pelo modelo, indicando quantas classificações positivas estavam realmente corretas. Já o recall mede a capacidade do modelo de identificar corretamente os exemplos positivos existentes no conjunto de dados. Dessa forma, precision está relacionada à redução de falsos positivos, enquanto recall está relacionada à redução de falsos negativos.

3) Em quais situações o f1-score é importante?

O f1-score é importante em problemas com classes desbalanceadas ou em situações onde precision e recall são igualmente relevantes. Como essa métrica combina ambas em um único valor, ela fornece uma avaliação mais equilibrada do desempenho do modelo, sendo muito utilizada em aplicações como detecção de fraudes, diagnósticos médicos e sistemas de segurança.

Implemente o rastreamento experimental utilizando MLflow.

## Devem ser registrados:

Ao final:

1. Qual experimento apresentou melhor desempenho?

**Solução**:

````python
import time
import pandas as pd
import mlflow

from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def run_experiment(
    group,
    X_train,
    X_val,
    y_train,
    y_val,
    activation,
    hidden_layers,
    learning_rate,
    max_iter,
    batch_size,
    seed
):
    with mlflow.start_run():

        mlflow.log_param("group", group)
        mlflow.log_param("activation", activation)
        mlflow.log_param("hidden_layers", hidden_layers)
        mlflow.log_param("learning_rate", learning_rate)
        mlflow.log_param("max_iter", max_iter)
        mlflow.log_param("batch_size", batch_size)

        model = MLPClassifier(
            hidden_layer_sizes=hidden_layers,
            activation=activation,
            learning_rate_init=learning_rate,
            max_iter=max_iter,
            batch_size=batch_size,
            random_state=seed
        )

        start_time = time.time()
        model.fit(X_train, y_train.ravel())
        training_time = time.time() - start_time

        y_pred = model.predict(X_val)

        accuracy = accuracy_score(y_val, y_pred)
        precision = precision_score(y_val, y_pred, average="weighted", zero_division=0)
        recall = recall_score(y_val, y_pred, average="weighted", zero_division=0)
        f1 = f1_score(y_val, y_pred, average="weighted", zero_division=0)

        mlflow.log_metric("accuracy", accuracy)
        mlflow.log_metric("precision", precision)
        mlflow.log_metric("recall", recall)
        mlflow.log_metric("f1_score", f1)
        mlflow.log_metric("training_time", training_time)

        return {
            "group": group,
            "activation": activation,
            "hidden_layers": hidden_layers,
            "learning_rate": learning_rate,
            "max_iter": max_iter,
            "batch_size": batch_size,
            "accuracy": accuracy,
            "precision": precision,
            "recall": recall,
            "f1_score": f1,
            "training_time": training_time
        }
````

````python
X_train, X_val, y_train, y_val = load_data(seed=42)

X_train_exp = X_train[:10000]
y_train_exp = y_train[:10000]

X_val_exp = X_val[:2000]
y_val_exp = y_val[:2000]
````

````python
results_q4 = []

experiments_q4 = [
    ("Q4", "relu", (64,), 0.001, 20, 128),
    ("Q4", "relu", (128,), 0.001, 20, 128),
    ("Q4", "relu", (128, 64), 0.001, 20, 128),
    ("Q4", "tanh", (128,), 0.001, 20, 128),
    ("Q4", "logistic", (128,), 0.001, 20, 128),
    ("Q4", "relu", (128,), 0.01, 20, 128)
]

for group, activation, hidden_layers, learning_rate, max_iter, batch_size in experiments_q4:
    result = run_experiment(
        group,
        X_train_exp,
        X_val_exp,
        y_train_exp,
        y_val_exp,
        activation,
        hidden_layers,
        learning_rate,
        max_iter,
        batch_size,
        seed=42
    )

    results_q4.append(result)

results_q4_df = pd.DataFrame(results_q4)
results_q4_df
````
